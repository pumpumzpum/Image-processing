### Pum1_re.jpg

### 2.1 Nearest Neighbor Interpolation

```python
image2 = cv2.imread('Pum1_re.jpg')
#image2 = cv2.resize(image2, (400, 600)) #width height
near_img = cv2.resize(image2, None, fx = 5 , fy = 5, interpolation = cv2.INTER_NEAREST)

plt.subplot(121)
plt.imshow(cv2.cvtColor(image2, cv2.COLOR_BGR2RGB))
plt.title('Original Image',fontsize = 30)
fig = plt.gcf()
fig.set_size_inches(18.5, 10.5)
plt.axis('off')

plt.subplot(122)
plt.imshow(cv2.cvtColor(near_img, cv2.COLOR_BGR2RGB))
plt.title('Adjust Image',fontsize = 30)
fig = plt.gcf()
fig.set_size_inches(14.5, 6.5)
plt.axis('off')
```

### คำอธิบาย : จากภาพเราจะเห็นได้ว่าการทำ Nearest Neighbor Interpolation เหมือนทำให้เกิดจุดพิกเซลที่เป็นสีเทาชัดเจนยิ่งขึ้น ถ้าดูจากตรงขอบของรูปบุคคล จะเห็นเป็นพิกเซลชัดขึ้น

### 2.2 Bilinear Interpolation

```python
image2 = cv2.imread('Pum1_re.jpg')
#image2 = cv2.resize(image2, (400, 600)) #width height
bilinear_img = cv2.resize(image2,None, fx = 5, fy = 5, interpolation = cv2.INTER_LINEAR)

plt.subplot(121)
plt.imshow(cv2.cvtColor(image2, cv2.COLOR_BGR2RGB))
plt.title('Original Image',fontsize = 30)
fig = plt.gcf()
fig.set_size_inches(18.5, 10.5)
plt.axis('off')

plt.subplot(122)
plt.imshow(cv2.cvtColor(bilinear_img, cv2.COLOR_BGR2RGB))
plt.title('Adjust Image',fontsize = 30)
fig = plt.gcf()
fig.set_size_inches(14.5, 6.5)
plt.axis('off')
```

### คำอธิบาย : เห็นได้จากภาพว่า  Bilinear Interpolation ทำให้เกิดภาพใหม่ที่ดู smooth ขึ้น และไม่มีขอบขึ้นชัดเจนเหมือนการทำ Nearest Neighbor Interpolation เพราะ เป็นการ weight เอา ภาพเดิม 4 จุดพิกเซลมา ทำให้เกิดภาพใหม่

### 2.3 Bicubic Interpolation

```python
image2 = cv2.imread('Pum1_re.jpg')
#image2 = cv2.resize(image2, (400, 600)) #width height
bicubic_img = cv2.resize(image2,None, fx = 5, fy = 5, interpolation = cv2.INTER_CUBIC)

plt.subplot(121)
plt.imshow(cv2.cvtColor(image2, cv2.COLOR_BGR2RGB))
plt.title('Original Image',fontsize = 30)
fig = plt.gcf()
fig.set_size_inches(18.5, 10.5)
plt.axis('off')

plt.subplot(122)
plt.imshow(cv2.cvtColor(bicubic_img, cv2.COLOR_BGR2RGB))
plt.title('Adjust Image',fontsize = 30)
fig = plt.gcf()
fig.set_size_inches(14.5, 6.5)
plt.axis('off')
```

### คำอธิบาย : การทำ Bicubic Interpolation ทำให้เราได้เห็นว่า ภาพใหม่ที่เกิดขึ้นนั้นมีรายละเอียดที่ใกล้เคียงกับวิธี Bilinear Interpolation แต่ภาพที่ได้จะมีบางจุดที่มีความสว่างมากกว่า เป็นเพราะว่าวิธีการ Bicubic มีวิธีการคิดที่ซับซ้อนมากกว่า

## 3. Program of gray level resolution

```python
import cv2 
import numpy as np

image3 = cv2.imread('Pum1.jpg', 0)
image3 = cv2.resize(image3, (250, 400))

k = 0
while k < 8:
    intensity_level = 2**k
    image_reduced = np.uint8(np.floor(np.double(image3)/intensity_level))
    norm_image = cv2.normalize(image_reduced, None, 0, 255, norm_type = cv2.NORM_MINMAX)
    gray_level = 2 ** (8-k)
    
    cv2.imshow('Gray Level'+str(gray_level),norm_image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    
    k = k+1
```
