### 2.1 Nearest Neighbor Interpolation

```python
image2 = cv2.imread('Pum1_re.jpg')
#image2 = cv2.resize(image2, (400, 600)) #width height
near_img = cv2.resize(image2, None, fx = 5 , fy = 5, interpolation = cv2.INTER_NEAREST)

plt.subplot(121)
plt.imshow(cv2.cvtColor(image2, cv2.COLOR_BGR2RGB))
plt.title('Original Image',fontsize = 30)
fig = plt.gcf()
fig.set_size_inches(18.5, 10.5)
plt.axis('off')

plt.subplot(122)
plt.imshow(cv2.cvtColor(near_img, cv2.COLOR_BGR2RGB))
plt.title('Adjust Image',fontsize = 30)
fig = plt.gcf()
fig.set_size_inches(14.5, 6.5)
plt.axis('off')
```

###  :  Nearest Neighbor Interpolation   

### 2.2 Bilinear Interpolation

```python
image2 = cv2.imread('Pum1_re.jpg')
#image2 = cv2.resize(image2, (400, 600)) #width height
bilinear_img = cv2.resize(image2,None, fx = 5, fy = 5, interpolation = cv2.INTER_LINEAR)

plt.subplot(121)
plt.imshow(cv2.cvtColor(image2, cv2.COLOR_BGR2RGB))
plt.title('Original Image',fontsize = 30)
fig = plt.gcf()
fig.set_size_inches(18.5, 10.5)
plt.axis('off')

plt.subplot(122)
plt.imshow(cv2.cvtColor(bilinear_img, cv2.COLOR_BGR2RGB))
plt.title('Adjust Image',fontsize = 30)
fig = plt.gcf()
fig.set_size_inches(14.5, 6.5)
plt.axis('off')
```

### 2.3 Bicubic Interpolation

## 3. Program of gray level resolution

```python
import cv2 
import numpy as np

image3 = cv2.imread('Pum1.jpg', 0)
image3 = cv2.resize(image3, (250, 400))

k = 0
while k < 8:
    intensity_level = 2**k
    image_reduced = np.uint8(np.floor(np.double(image3)/intensity_level))
    norm_image = cv2.normalize(image_reduced, None, 0, 255, norm_type = cv2.NORM_MINMAX)
    gray_level = 2 ** (8-k)
    
    cv2.imshow('Gray Level'+str(gray_level),norm_image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    
    k = k+1
```
